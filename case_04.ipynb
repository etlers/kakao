```
words = ["frodo", "front", "frost", "frozen", "frame", "kakao"]
queries = ["fro??", "????o", "fr???", "fro???", "pro?"]
# [3, 2, 4, 1, 0]
```

```
def solution(words, queries):
    # 질의어에 대한 개수 저장을 위한 딕셔너리
    answer_dict = {}

    # 입력받은 문자열을 리스트로 변환
    def string_to_list(in_string):
        list_value = []
        for ii in range(len(in_string)):
            list_value.append(in_string[ii:ii+1])
        return list_value

    # 문자에서 질의어 찾기
    def compare_charater_in_word(querie, word):
        list_querie = string_to_list(querie)
        list_word = string_to_list(word)
        for ii in range(len(list_querie)):
            # 와일드 카드는 비교에서 제외
            if list_querie[ii] == "?": continue
            # 다른지만 확인
            if list_querie[ii] != list_word[ii]:
                return False
        return True

    # 질의어를 먼저 추출
    for querie in queries:
        match_cnt = 0
        # 검색할 단어를 추출
        for word in words:
            if len(word) == len(querie):
                # 와일드 카드를 제외하고 위치에 문자가 일치했다면
                if compare_charater_in_word(querie, word) == True:
                    match_cnt += 1
        # 딕셔너리에 질의어와 일치한 개수 저장
        answer_dict[querie] = match_cnt

    # 리스트로 딕셔너리의 값만 전송
    return list(answer_dict.values())
```

```
solution(words, queries)
```
