1. 입력이 빈 문자열인 경우, 빈 문자열을 반환합니다.
2. 문자열 w를 두 "균형잡힌 괄호 문자열" u, v로 분리합니다. 단, u는 "균형잡힌 괄호 문자열"로 더 이상 분리할 수 없어야 하며, v는 빈 문자열이 될 수 있습니다.
3. 문자열 u가 "올바른 괄호 문자열" 이라면 문자열 v에 대해 1단계부터 다시 수행합니다.
  3-1. 수행한 결과 문자열을 u에 이어 붙인 후 반환합니다.
4. 문자열 u가 "올바른 괄호 문자열"이 아니라면 아래 과정을 수행합니다.
  4-1. 빈 문자열에 첫 번째 문자로 '('를 붙입니다.
  4-2. 문자열 v에 대해 1단계부터 재귀적으로 수행한 결과 문자열을 이어 붙입니다.
  4-3. ')'를 다시 붙입니다.
  4-4. u의 첫 번째와 마지막 문자를 제거하고, 나머지 문자열의 괄호 방향을 뒤집어서 뒤에 붙입니다.
  4-5. 생성된 문자열을 반환합니다.

")(", "()))((()", "(()())"

```
def solution(w):
    global balanced_string
    balanced_string = ""

    if len(w.strip()) == 0:
        return ""           

    def make_list_char(in_string):
        ii = 0
        list_char = []
        while ii < len(in_string):
            list_char.append(in_string[ii:ii+1])
            ii += 1
        return list_char
    
    def change_bracket(in_string):
        list_char = make_list_char(in_string)
        changed_string = []
        for idx in list_char:
            if idx == "(":
                changed_string.append(")")
            else:
                changed_string.append("(")
        return "".join(changed_string)

    def check_valid(in_string):        
        ix = 1
        open_cnt = 0
        close_cnt = 0
        invalid_yn = False
        list_char = make_list_char(in_string)
        for idx in list_char:
            # 처음에 닫힌 괄호가 나오면 완전한 괄호가 될 수 없음
            if (ix == 1) and (idx == ")"):
                return in_string
            if idx == "(":
                open_cnt += 1
            else:
                close_cnt += 1
            ix += 1
            
            if close_cnt > open_cnt:
                invalid_yn = True

        if invalid_yn:
            return in_string
        else:
            return "valid"

    def split_uv_by_string(u_string):   
        global balanced_string

        if len(u_string) == 2:
            return "()"
        temp_char = ""
        ix = 1
        list_char = make_list_char(u_string)
        for idx in list_char:
            if temp_char != "":
                if temp_char == idx:
                    ix += 1
                else:
                    break
            temp_char = idx

        u = ""
        v = ""
        for idx in range(ix):
            u += temp_char
        for idx in range(ix):
            if temp_char == "(":
                u += ")"
            else:
                u += "("
        v = u_string[ix*2:]
        # U가 올바른지 확인. 올바르다면 V로 재귀함수 호출
        if check_valid(u) == "valid":
            balanced_string += u + split_uv_by_string(v)
            return balanced_string
        else:
            return balanced_string + "(" + change_bracket(u[1:len(u)-1]) + ")" + v
    # 올바른 괄호이면 메세지 던지고 빠져나감
    if check_valid(w) == "valid":
        return w
    # 균형잡힌 괄호 생성
    answer = split_uv_by_string(w)
    return answer
```

```
# 첫번째 괄호와 두번째 괄호가 다른지 확인. 다르면 한 쌍으로, 같으면 다른게 나올 때까지
s = "(()())"
# "("로 시작하면 올바른 문자열
ii = 0
list_char = []
while ii < len(s):
    list_char.append(s[ii:ii+1])
    ii += 1
temp_char = ""
ix = 1
for idx in list_char:
    if temp_char != "":
        if temp_char == idx:
            ix += 1
        else:
            break
    temp_char = idx

u = ""
v = ""
for idx in range(ix):
    u += temp_char
for idx in range(ix):
    if temp_char == "(":
        u += ")"
    else:
        u += "("
v = s.replace(u, "")
print(u, v)
```
