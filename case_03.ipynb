```
def solution(key, lock):

    # 행열의 개수
    r_cnt = np.size(lock, 0)
    list_direction = ["R", "D", "L", "U"]

    # 입력받은 배열을 최종 검수할 배열의 크기에 맞추기
    def make_new_array(in_arr):
        new_arr = np.zeros([r_cnt,r_cnt]).astype(np.int64)
        arr_size = np.size(lock, 0)
        for ix in range(arr_size):
            for iy in range(arr_size):
                if in_arr[ix,iy] == 1:
                    new_arr[ix,iy] = 1
        return new_arr

    # 더해서 모두 "1"이면 정답
    def check_match(in_arr):
        # 매치에 필요한 개수
        amswer = lock + in_arr
        # 모두 "1"이 존재하면 정답이 아님
        for ix in range(r_cnt):
            for iy in range(r_cnt):
                if amswer[ix,iy] == 0:
                    return False
        return True

    # 우측으로 90도 회전
    def rotation_array_to_right(in_arr):
        # 행, 열의 최대 위치 값
        r_max = r_cnt - 1
        # 회전한 배열 생성
        arr = np.zeros([r_cnt,r_cnt])
        # 행 체크
        for ix in range(r_cnt):
            if ix == 0:
                ry = r_max
            elif ix == r_max:
                ry = 0            
            else:
                ry = r_max - ix
            # 열 체크
            for ii in range(r_cnt):
                arr[ii, ry] = in_arr[ix, ii]
        return arr.astype(np.int64)      

    # 네 방위로 이동
    def move_four_direction(dir, in_arr):
        ic = 1
        while ic < r_cnt:
            new_arr = np.zeros([r_cnt,r_cnt]).astype(np.int64)
            if dir == "R":
                for ix in range(r_cnt):
                    for iy in range(r_cnt-ic):
                        if in_arr[ix,iy] == 1:                        
                            new_arr[ix,iy+ic] = in_arr[ix,iy]
            elif dir == "D":
                for ix in range(r_cnt-ic):
                    for iy in range(r_cnt):
                        if in_arr[ix,iy] == 1:
                            new_arr[ix+ic,iy] = in_arr[ix,iy]
            if dir == "L":
                for ix in range(r_cnt):
                    for iy in range(r_cnt):                    
                        if (iy-ic >= 0) and (in_arr[ix,iy] == 1):
                            new_arr[ix,iy-ic] = in_arr[ix,iy]                
            elif dir == "U":
                for ix in range(r_cnt):
                    for iy in range(r_cnt):
                        if (ix-ic >= 0) and (in_arr[ix,iy] == 1):
                            new_arr[ix-ic,iy] = in_arr[ix,iy]
            ic += 1
            # 정답이 있다면 종료
            if check_match(new_arr) == True:
                return True
        return False

    # 360도 4방위 확인
    def four_direction_and_full_rotation():
        chg_arr = key
        for _ in range(3):
            # 90도 회전
            chg_arr = rotation_array_to_right(chg_arr)
            # 회전한 배열로 재배열
            check_arr = make_new_array(chg_arr)
            print(check_arr)
            # 정답이면 빠져나감
            if check_match(check_arr) == True:
                return "true"
            # 네 방위로 돌면서 확인            
            for dir in list_direction:
                if move_four_direction(dir, check_arr) == True:
                    return "true"
        return "false"

    # 처음은 있는 그대로 lock 배열에 맞춰 재생성 후 확인
    check_arr = make_new_array(key)
    # 정답이면 빠져나감
    if check_match(check_arr) == True:
        return "true"
    
    # 360도 4방위로 확인
    answer = four_direction_and_full_rotation()
    return answer
```
